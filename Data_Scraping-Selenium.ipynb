### Problem Statement - Finding Information Technology open positions in Hyderabad location

```python
#Identify Key Tech Hires in the Hyderabad area -
#Companies (both international and local) competing for local talent pool.
#Our aim is to Identify the key Companies in the market who are hiring in Hyderabad area.
#When you have those, we want to know how many positions they have and what are the Key skills they are hiring for.
#Your output file may look like this
#Name of the company, Number of Positions , Segmentation, Skills they are looking for
#============================================================
#Amazon, 2300, “Information technology” , “Tableau, R, Python, Java, Salesforce”
#Flipkart, 1300, “Information technology” , “Java, API, REST API, Integration, C++, Python”
#Microsoft, 8600, “information technology”, “.net, Machine Learning, AI, Tableau”
```

```python
from selenium import webdriver
import chromedriver_binary
import time
import pandas as pd 
import datetime
import urllib.request
import warnings
```

#### Scraping naukri.com...

```python
urllink = "https://www.naukri.com/information-technology-jobs-in-hyderabad-secunderabad" # URL to scrape
warnings.filterwarnings('ignore') # to ignore any warnings
pd.set_option('display.max_colwidth', 100) # to display dataframe column data upto 100 characters
```

#### Ask for chrome driver location

```python
#driverlocation = "D:/Softwares/chromedriver_win32/chromedriver.exe"
driverlocation = input("Enter the location of Chrome Driver: ")
```

#### Lets get the input from user to scrape the number of pages. The higher the number, the longer the time it takes to scrape...

```python
pages_to_scrape = input("Enter the number(int) of pages to scrape: ")

# Exception handling for the input

try:
    pages_to_scrape = int(pages_to_scrape)
except ValueError:
    try:
        pages_to_scrape = int(round(float(pages_to_scrape)))
        print("Input is a float  number so it's interger value will be considered i.e. ", pages_to_scrape)
    except ValueError:
        print("Input is not a number. No need to proceed...")
```

#### Define empty lists for required attributes and then populate the same in a loop...

```python
driver = webdriver.Chrome(driverlocation)

company_name = [] # empty list to be populated with company names
designation = [] # empty list to be populated with designations
skills = [] # empty list to be populated with skills

jobs_per_page = 20  # There are 20 job postings on any naukri.com page

for i in range(0,pages_to_scrape):   
    if i==0:
        urlToBrowse = urllink
    else:
        urlToBrowse = urllink + "-" + str(i+1)  # adding page number with '-' from 2nd page onwards
        
    driver.get(urlToBrowse)
    
    for j in range(1,jobs_per_page+1):
        company_name.append(driver.find_element_by_xpath( \
            '//*[@id="root"]/div[4]/div[2]/section[2]/div[2]/article['+str(j)+']/div[1]/div/div/a[1]').text)
        designation.append(driver.find_element_by_xpath( \
            '//*[@id="root"]/div[4]/div[2]/section[2]/div[2]/article['+str(j)+']/div[1]/div/a').text)
        skills.append(driver.find_element_by_xpath( \
            '//*[@id="root"]/div[4]/div[2]/section[2]/div[2]/article['+str(j)+']/ul').text)
        time.sleep(1)
    time.sleep(5)
    
driver.close() # Closing the driver

#print(company_name)
#print(designation)
#print(skills)
```

#### Create a dataframe with columns as the lists populated above

```python
df_jobs = pd.DataFrame()
df_jobs["company_name"] = company_name
df_jobs["designation"] = designation
df_jobs["skills"] = skills
df_jobs
```

#### Groupby the dataframe by 'company_name' and concatenate 'designation' and 'skills'. Use delimiter as '|-|' to avoid any occurence in data

```python
df_jobs_agg = df_jobs.groupby(['company_name'],as_index=False)['designation','skills'].agg('|-|'.join)
df_jobs_agg
```

#### Add a new attribute for number of positions. Populate with the number of values in aggregated 'designation' attribute (separated by '|-|')

```python
df_jobs_agg["number_of_positions"] =  df_jobs_agg['designation'].apply(lambda i: len(i.split('|-|')))
df_jobs_agg
```

#### Replace newline with comma in 'skills' attribute

```python
df_jobs_agg["skills"] = df_jobs_agg["skills"].str.replace("\n",",")
df_jobs_agg
```

#### Add a new attribute called 'Segmentation'. Populate with "Information Technology"

```python
df_jobs_agg["Segmentation"] = "Information Technology"
df_jobs_agg
```

#### Create a final dataframe. Drop 'designation' attribute as it's not required

```python
final_df = pd.DataFrame()
final_df["Name of the company"] = df_jobs_agg["company_name"]
final_df["Number of Positions"] = df_jobs_agg["number_of_positions"]
final_df["Segmentation"] = df_jobs_agg["Segmentation"]
final_df["Skills they are looking for"] = df_jobs_agg["skills"]
final_df
```

#### Write the dataframe to CSV file

```python
final_df.to_csv('./Hyderabad_Anuroop_Ajmera.csv', index=False)
```
